The output off `image_to_string`is the full text extracted.

`image_to_boxes` returns the bounding boxes of the characters.

```python
ticket_data = tess.image_to_data(ticket)
print(type(ticket_data))
print(ticket_data)
```

```python
import io
ticket_df = pd.read_csv(io.StringIO(ticket_data), sep='\s+')
```

```python
# display a sample
ticket_df.sample(5)
```

We see several -1 confidence values in the output of `image_to_boxes` with an empty string as the character. We can filter these out to get the bounding boxes of the characters with a confidence value greater than -1.

```python
# filter
ticket_df = ticket_df[ticket_df["conf"] != -1]
ticket_df.sample(5)
```

```python
from PIL import ImageDraw, ImageFont
def draw_box(image, left, top, width, height, conf, text):
    # copy of the original image to avoid overwritting
    image_with_box = image.copy()
    
    # create drawing
    draw = ImageDraw.Draw(image_with_box)
    
    # coordinates of the bounding box
    right = left + width
    bottom = top + height
    
    # draw the red rectangle
    draw.rectangle([(left, top), (right, bottom)], outline='red', width=2)
    
    # font
    try:
        font = ImageFont.truetype("/System/Library/Fonts/Arial.ttf", 20)
    except:
        font = ImageFont.load_default()
    
    # draw text above the box
    text_to_draw = f"{text} ({conf:.2f})"
    draw.text((left, top-25), text_to_draw, fill='red', font=font)
    
    return image_with_box
```

```python
image_draw = draw_box(ticket, 100,100,100,100,99,"test")
display(image_draw)
```

```python
def draw_boxes_from_data(image, extracted_data):
    image_draw = image.copy()
    for box in extracted_data.itertuples():
        left = box.left
        top = box.top
        width = box.width
        height = box.height
        conf = box.conf/100
        text = box.text
        image_draw = draw_box(image_draw,left,top,width,height,conf,text)
    return image_draw       
```

```python
ticket_boxes_draw = draw_boxes_from_data(ticket, ticket_df)
display(ticket_boxes_draw)
```
